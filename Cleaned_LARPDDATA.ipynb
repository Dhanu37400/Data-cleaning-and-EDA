```python

ct = pd.crosstab(data['LARP PROCEDURE'], data['Category'])
ax = ct.plot(
    kind='bar',
    stacked=True,
    figsize=(10,6),
    colormap='Set2'
)

plt.title("LARP Procedure vs Incident Category", fontsize=14)
plt.xlabel("LARP Procedure")
plt.ylabel("Number of Incidents")

plt.legend(title="Category")

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```
