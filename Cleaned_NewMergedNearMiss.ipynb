```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    x='Severity of incident',
    data=data,
    order=data['Severity of incident'].value_counts().index
)
for p in ax.patches:
    ax.annotate(int(p.get_height()),
                (p.get_x() + 0.3, p.get_height() + 5))

plt.title("Severity Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))

ax = sns.countplot(
    x='Probability',
    data=data,
    order=data['Probability'].value_counts().index
)
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width()/2, p.get_height()),
        ha='center', va='bottom', fontsize=10
    )
total = len(data)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(
        percentage,
        (p.get_x() + p.get_width()/2, p.get_height()/2),
        ha='center', va='center', color='white', fontsize=9, weight='bold'
    )
plt.title("Probability Distribution (Count + %)", fontsize=14, weight='bold')
plt.xlabel("Probability Level", fontsize=12)
plt.ylabel("Number of Incidents", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
order = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

ax = sns.countplot(
    x='Probability',
    data=data,
    order=order,
    palette='viridis'
)

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(
            f'{int(height)}',
            (p.get_x() + p.get_width()/2, height),
            ha='center', va='bottom', fontsize=10
        )

total = len(data)
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        percentage = f'{100 * height / total:.1f}%'
        ax.annotate(
            percentage,
            (p.get_x() + p.get_width()/2, height/2),
            ha='center', va='center',
            color='white', fontsize=9, weight='bold'
        )

plt.title("Probability Distribution (Count + Percentage)", fontsize=14, weight='bold')
plt.xlabel("Probability Level", fontsize=12)
plt.ylabel("Number of Incidents", fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
