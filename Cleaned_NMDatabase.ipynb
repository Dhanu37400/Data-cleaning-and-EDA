```python
import re

def clean_noise(x):
    if isinstance(x, str):
        x = re.sub(r'Unnamed:\s*\d+', '', x)
        x = re.sub(r'dtype:\s*\w+', '', x)
        x = re.sub(r'Name:\s*\d+', '', x)
    return x

data['Details_of_potential_loss_Category'] = data['Details_of_potential_loss_Category'].apply(clean_noise)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Severity Of Incident', data=data)
plt.title("Incident Count by Severity")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Type Of Ship', data=data)
plt.xticks(rotation=45)
plt.title("Incidents by Ship Type")
plt.show()
```

```python
data['Date Of Report'] = pd.to_datetime(data['Date Of Report'])

data['Year'] = data['Date Of Report'].dt.year

sns.countplot(x='Year', data=data)
plt.title("Year-wise Incident Trend")
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax = data['Location'].value_counts().head(10).plot(
    kind='bar',
    edgecolor='black'
)
plt.title("Top 10 Incident Locations", fontsize=14)
plt.xlabel("Location")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    x='Type Of Ship',
    hue='Severity Of Incident',
    data=data
)

plt.xticks(rotation=45)
plt.title("Severity Distribution by Ship Type")
plt.show()
```

```python
data['Month'] = data['Date Of Report'].dt.month

sns.countplot(x='Month', data=data)
plt.title("Monthly Incident Trend")
plt.show()
```
